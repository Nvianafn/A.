3.SPK SAW

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

def normalize(matrix, types):
    norm_matrix = np.zeros_like(matrix, dtype=float)
    for j in range(matrix.shape[1]):
        if types[j] == 'benefit':
            norm_matrix[:, j] = matrix[:, j] / np.max(matrix[:, j])
        elif types[j] == 'cost':
            norm_matrix[:, j] = np.min(matrix[:, j]) / matrix[:, j]
    return norm_matrix

def saw_method(candidates, criteria, weights, values, types):
    values = np.array(values, dtype=float)
    weights = np.array(weights, dtype=float)

    # Normalisasi matriks
    norm_values = normalize(values, types)

    # Menghitung skor SAW
    scores = np.dot(norm_values, weights)

    # Hasil dalam bentuk tabel
    results = pd.DataFrame({'Kandidat': candidates, 'Skor': scores})
    results = results.sort_values(by='Skor', ascending=False)

    return results

def main():
    # Input data
    n_candidates = int(input("Masukkan jumlah kandidat: "))
    n_criteria = int(input("Masukkan jumlah kriteria: "))

    candidates = [input(f"Nama kandidat {i+1}: ") for i in range(n_candidates)]
    criteria = [input(f"Nama kriteria {j+1}: ") for j in range(n_criteria)]

    weights = list(map(float, input(f"Masukkan bobot untuk setiap kriteria ({n_criteria} nilai, dipisahkan spasi): ").split()))

    types = [input(f"Tipe kriteria '{criteria[j]}' (benefit/cost): ") for j in range(n_criteria)]

    values = []
    for i in range(n_candidates):
        print(f"Masukkan nilai untuk {candidates[i]} (pisahkan dengan spasi, {n_criteria} nilai):")
        values.append(list(map(float, input().split())))

    # Perhitungan SAW
    results = saw_method(candidates, criteria, weights, values, types)

    # Menampilkan hasil dalam tabel
    print("\nHasil Perhitungan SAW:")
    print(results)

    # Plot hasil dalam bentuk grafik
    plt.figure(figsize=(8, 5))
    plt.bar(results['Kandidat'], results['Skor'], color='skyblue')
    plt.xlabel("Kandidat")
    plt.ylabel("Skor Akhir")
    plt.title("Hasil Perhitungan SAW")
    plt.xticks(rotation=45)
    plt.show()

if __name__ == "__main__":
    main()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx

def expected_utility(p, U_10000, U_0):
    return p * U_10000 + (1 - p) * U_0

# Diketahui
p = 0.8  # Probabilitas sukses investasi
U_10000 = 1  # Utilitas uang $10,000
U_0 = 0  # Utilitas uang $0

# Hitung utilitas uang $5000
U_5000 = expected_utility(p, U_10000, U_0)

# Langkah-langkah perhitungan
steps = [
    f"1. Diketahui: P(sukses) = {p}, U($10,000) = {U_10000}, U($0) = {U_0}",
    "2. Gunakan rumus Expected Utility:",
    "   U($5000) = p * U($10,000) + (1 - p) * U($0)",
    f"3. Substitusi nilai: U($5000) = ({p} * {U_10000}) + ({1-p} * {U_0})",
    f"4. Hasil perhitungan: U($5000) = {U_5000}"
]

# Cetak langkah-langkah perhitungan
for step in steps:
    print(step)

# Membuat pohon keputusan
G = nx.DiGraph()
G.add_node("Mulai")
G.add_node("Investasi Real Estate")
G.add_node("Simpan di Bank")
G.add_node("Sukses ($10,000)")
G.add_node("Gagal ($0)")
G.add_node("Keuntungan Bank ($5,000)")
# Tambahkan edges
G.add_edges_from([
    ("Mulai", "Investasi Real Estate"),
    ("Mulai", "Simpan di Bank"),
    ("Investasi Real Estate", "Sukses ($10,000)", {'prob': p}),
    ("Investasi Real Estate", "Gagal ($0)", {'prob': 1 - p}),
    ("Simpan di Bank", "Keuntungan Bank ($5,000)", {'prob': 1})
])

# Posisi node dalam plot
pos = {
    "Mulai": (0, 2),
    "Investasi Real Estate": (-1, 1),
    "Simpan di Bank": (1, 1),
    "Sukses ($10,000)": (-1.5, 0),
    "Gagal ($0)": (-0.5, 0),
    "Keuntungan Bank ($5,000)": (1, 0)
}

# Plot pohon keputusan
plt.figure(figsize=(8, 5))
nx.draw(G, pos, with_labels=True, node_size=3000, node_color='lightblue', edge_color='gray', font_size=9)
edge_labels = {edge: f"p={G.edges[edge]['prob']:.1f}" for edge in G.edges if 'prob' in G.edges[edge]}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=9)
plt.title("Pohon Keputusan untuk Investasi")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Fungsi keanggotaan linear
def turun(x, x_min, x_max):
    return np.clip((x_max - x) / (x_max - x_min), 0, 1) # Indented this line
def naik(x, x_min, x_max):
    return np.clip((x - x_min) / (x_max - x_min), 0, 1) # Indented this line
def sedikit(x, x_min, x_max):
    return turun(x, x_min, x_max) # Indented this line
def banyak(x, x_min, x_max):
    return naik(x, x_min, x_max) # Indented this line
def berkurang(, min_output, max_output):
    return max_output -  * (max_output - min_output) # Indented this line
def bertambah(, min_output, max_output):
    return  * (max_output - min_output) + min_output # Indented this line
# Parameter dari soal
permintaan_min = 1000
permintaan_max = 5000
persediaan_min = 100
persediaan_max = 600
produksi_min = 2000
produksi_max = 7000
# Input dari user
permintaan = float(input("Masukkan jumlah permintaan (kemasan): "))  # contoh: 4000
persediaan = float(input("Masukkan jumlah persediaan (kemasan): "))  # contoh: 300
# Hitung derajat keanggotaan
_perm_turun = turun(permintaan, permintaan_min, permintaan_max)
_perm_naik = naik(permintaan, permintaan_min, permintaan_max)
_pers_sedikit = sedikit(persediaan, persediaan_min, persediaan_max)
_pers_banyak = banyak(persediaan, persediaan_min, persediaan_max)
# Output perhitungan derajat keanggotaan
print("\n=== DERAJAT KEANGGOTAAN ===")
print(f"Permintaan Turun: {_perm_turun:.4f}")
print(f"Permintaan Naik : {_perm_naik:.4f}")
print(f"Persediaan Sedikit: {_pers_sedikit:.4f}")
print(f"Persediaan Banyak : {_pers_banyak:.4f}")
# Aturan fuzzy
print("\n=== PERHITUNGAN ATURAN ===")
# Rule 1
1 = min(_perm_turun, _pers_banyak)
z1 = berkurang(1, produksi_min, produksi_max)
print(f"Rule 1: IF permintaan TURUN and persediaan BANYAK THEN produksi BERKURANG")
print(f"  1 = min({_perm_turun:.4f}, {_pers_banyak:.4f}) = {1:.4f}")
print(f"  z1 = produksi berkurang = {z1:.2f} kemasan")
# Rule 2
2 = min(_perm_turun, _pers_sedikit)
z2 = berkurang(2, produksi_min, produksi_max)
print(f"Rule 2: IF permintaan TURUN and persediaan SEDIKIT THEN produksi BERKURANG")
print(f"  2 = min({_perm_turun:.4f}, {_pers_sedikit:.4f}) = {2:.4f}")
print(f"  z2 = produksi berkurang = {z2:.2f} kemasan")
# Rule 3
3 = min(_perm_naik, _pers_banyak)
z3 = bertambah(3, produksi_min, produksi_max)
print(f"Rule 3: IF permintaan NAIK and persediaan BANYAK THEN produksi BERTAMBAH")
print(f"  3 = min({_perm_naik:.4f}, {_pers_banyak:.4f}) = {3:.4f}")
print(f"  z3 = produksi bertambah = {z3:.2f} kemasan")
# Rule 4
4 = min(_perm_naik, _pers_sedikit)
z4 = bertambah(4, produksi_min, produksi_max)
print(f"Rule 4: IF permintaan NAIK and persediaan SEDIKIT THEN produksi BERTAMBAH")
print(f"  4 = min({_perm_naik:.4f}, {_pers_sedikit:.4f}) = {4:.4f}")
print(f"  z4 = produksi bertambah = {z4:.2f} kemasan")
# Defuzzifikasi (Weighted Average)
numerator = 1 * z1 + 2 * z2 + 3 * z3 + 4 * z4
denominator = 1 + 2 + 3 + 4
print("\n=== DEFUZZIFIKASI ===")
print(f"Numerator   = {numerator:.4f}")
print(f"Denominator = {denominator:.4f}")
produksi_final = numerator / denominator if denominator != 0 else 0
print(f"\nJumlah produksi yang disarankan: {produksi_final:.2f} kemasan")
# ---------------------------
# GRAFIK KEANGGOTAAN
# ---------------------------
x_permintaan = np.linspace(permintaan_min, permintaan_max, 100)
x_persediaan = np.linspace(persediaan_min, persediaan_max, 100)
plt.figure(figsize=(12, 5))
# Grafik Permintaan
plt.subplot(1, 2, 1)
plt.plot(x_permintaan, [turun(x, permintaan_min, permintaan_max) for x in x_permintaan],
label='Permintaan Turun')
plt.plot(x_permintaan, [naik(x, permintaan_min, permintaan_max) for x in x_permintaan],
label='Permintaan Naik')
plt.axvline(x=permintaan, color='red', linestyle='--', label=f'Input = {permintaan}')
plt.title("Fungsi Keanggotaan Permintaan")
plt.xlabel("Permintaan (kemasan)")
plt.ylabel("Derajat")
plt.legend()
plt.grid(True)
# Grafik Persediaan
plt.subplot(1, 2, 2)
plt.plot(x_persediaan, [sedikit(x, persediaan_min, persediaan_max) for x in x_persediaan],
label='Persediaan Sedikit')
plt.plot(x_persediaan, [banyak(x, persediaan_min, persediaan_max) for x in x_persediaan],
label='Persediaan Banyak')
plt.axvline(x=persediaan, color='red', linestyle='--', label=f'Input = {persediaan}')
plt.title("Fungsi Keanggotaan Persediaan")
plt.xlabel("Persediaan (kemasan)")
plt.ylabel("Derajat")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Fungsi keanggotaan input
def turun(x, xmin, xmax):
    return np.clip((xmax - x) / (xmax - xmin), 0, 1)

def naik(x, xmin, xmax):
    return np.clip((x - xmin) / (xmax - xmin), 0, 1)

# Fungsi keanggotaan output fuzzy
def produksi_berkurang(x):
    return turun(x, produksi_min, produksi_max)

def produksi_bertambah(x):
    return naik(x, produksi_min, produksi_max)

# Parameter
permintaan_min, permintaan_max = 1000, 5000
persediaan_min, persediaan_max = 100, 600
produksi_min, produksi_max = 2000, 7000

# Input user
permintaan = float(input("Masukkan jumlah permintaan: "))
persediaan = float(input("Masukkan jumlah persediaan: "))

# Derajat keanggotaan input
_perm_turun = turun(permintaan, permintaan_min, permintaan_max)
_perm_naik = naik(permintaan, permintaan_min, permintaan_max)
_pers_sedikit = turun(persediaan, persediaan_min, persediaan_max)
_pers_banyak = naik(persediaan, persediaan_min, persediaan_max)

print("\n=== DERAJAT KEANGGOTAAN ===")
print(f"Permintaan Turun : {_perm_turun:.4f}")
print(f"Permintaan Naik  : {_perm_naik:.4f}")
print(f"Persediaan Sedikit: {_pers_sedikit:.4f}")
print(f"Persediaan Banyak : {_pers_banyak:.4f}")

# Inferensi fuzzy Mamdani
1 = min(_perm_turun, _pers_banyak)
2 = min(_perm_turun, _pers_sedikit)
3 = min(_perm_naik, _pers_banyak)
4 = min(_perm_naik, _pers_sedikit)

# Domain output
x_output = np.linspace(produksi_min, produksi_max, 1000)

rule1 = np.fmin(1, produksi_berkurang(x_output))
rule2 = np.fmin(2, produksi_berkurang(x_output))
rule3 = np.fmin(3, produksi_bertambah(x_output))
rule4 = np.fmin(4, produksi_bertambah(x_output))

# Gabungan output
output_aggregated = np.fmax(rule1, np.fmax(rule2, np.fmax(rule3, rule4)))

# Defuzzifikasi (centroid)
if np.sum(output_aggregated) != 0:
    produksi_final = np.sum(output_aggregated * x_output) / np.sum(output_aggregated)
else:
    produksi_final = 0

print(f"\nJumlah produksi yang disarankan (Mamdani): {produksi_final:.2f} kemasan")

# ============================
# GABUNGKAN VISUALISASI
# ============================
fig, axs = plt.subplots(1, 3, figsize=(18, 5))

# Permintaan
x_permintaan = np.linspace(permintaan_min, permintaan_max, 100)
axs[0].plot(x_permintaan, [turun(x, permintaan_min, permintaan_max) for x in x_permintaan], label="Permintaan Turun")
axs[0].plot(x_permintaan, [naik(x, permintaan_min, permintaan_max) for x in x_permintaan], label="Permintaan Naik")
axs[0].axvline(x=permintaan, color='red', linestyle='--', label=f'Input = {permintaan}')
axs[0].set_title("Fungsi Keanggotaan Permintaan")
axs[0].set_xlabel("Permintaan (kemasan)")
axs[0].set_ylabel("Derajat")
axs[0].legend()
axs[0].grid(True)

# Persediaan
x_persediaan = np.linspace(persediaan_min, persediaan_max, 100)
axs[1].plot(x_persediaan, [turun(x, persediaan_min, persediaan_max) for x in x_persediaan], label="Persediaan Sedikit")
axs[1].plot(x_persediaan, [naik(x, persediaan_min, persediaan_max) for x in x_persediaan], label="Persediaan Banyak")
axs[1].axvline(x=persediaan, color='red', linestyle='--', label=f'Input = {persediaan}')
axs[1].set_title("Fungsi Keanggotaan Persediaan")
axs[1].set_xlabel("Persediaan (kemasan)")
axs[1].legend()
axs[1].grid(True)

# Output Produksi
axs[2].plot(x_output, produksi_berkurang(x_output), 'b--', label='Produksi Berkurang')
axs[2].plot(x_output, produksi_bertambah(x_output), 'g--', label='Produksi Bertambah')
axs[2].fill_between(x_output, output_aggregated, color='orange', alpha=0.6, label='Gabungan Output')
axs[2].axvline(produksi_final, color='red', linestyle='--', label=f'Defuzzifikasi = {produksi_final:.2f}')
axs[2].set_title("Output Produksi (Mamdani)")
axs[2].set_xlabel("Produksi (kemasan)")
axs[2].legend()
axs[2].grid(True)

plt.tight_layout()
plt.show()
```

1.   suhu ruangan (C)
2.   tingkat kelembaban udara (%)


---
Variabel dan Rentang Nilai:  Input 1: Suhu (C)  Rendah: 18  24  Tinggi: 22  30  Input 2: Kelembapan (%)  Kering: 30  50  Lembap: 40  70  Output: Kecepatan Kipas (Fan Speed dalam %)  Lambat: 0  50  Cepat: 40  100


---


Aturan Fuzzy (Rule Base):


1.   IF Suhu rendah AND Kelembapan kering  Fan speed lambat
2.   IF Suhu rendah AND Kelembapan lembap  Fan speed cepat
3.   IF Suhu tinggi AND Kelembapan kering  Fan speed cepat
4.   IF Suhu tinggi AND Kelembapan lembap  Fan speed cepat


---

 Pertanyaan: Jika diketahui:  Suhu ruangan saat ini adalah 26C  Kelembapan udara adalah 45% Gunakan metode fuzzy Mamdani untuk menentukan kecepatan kipas (%) yang direkomendasikan. Tampilkan grafik fungsi keanggotaan untuk:  Suhu  Kelembapan  Fan speed (beserta hasil defuzzifikasi)

```python
import numpy as np
import matplotlib.pyplot as plt

# Cell 1 - Fungsi Keanggotaan
def turun(x, x_min, x_max):
    return np.clip((x_max - x) / (x_max - x_min), 0, 1)

def naik(x, x_min, x_max):
    return np.clip((x - x_min) / (x_max - x_min), 0, 1)

def cahaya_rendah(x): return turun(x, 0, 5000)
def cahaya_tinggi(x): return naik(x, 3000, 10000)

def tanah_kering(x): return turun(x, 0, 50)
def tanah_basah(x): return naik(x, 30, 100)

# Cell 2 - Fungsi Output dan Aturan
def penyiraman_pendek(): return 0, 10
def penyiraman_sedang(): return 5, 20
def penyiraman_panjang(): return 15, 30

def berkurang(, z_min, z_max): return z_max -  * (z_max - z_min)
def bertambah(, z_min, z_max): return  * (z_max - z_min) + z_min

# Cell 3 - Input dan Derajat Keanggotaan
# Input intensitas cahaya dan kelembapan tanah
cahaya = 6200
kelembapan = 40

# Hitung derajat keanggotaan
_cahaya_rendah = cahaya_rendah(cahaya)
_cahaya_tinggi = cahaya_tinggi(cahaya)
_kering = tanah_kering(kelembapan)
_basah = tanah_basah(kelembapan)

# Cetak hasil
print("=== DERAJAT KEANGGOTAAN ===")
print(f"Cahaya Rendah : {_cahaya_rendah:.4f}")
print(f"Cahaya Tinggi : {_cahaya_tinggi:.4f}")
print(f"Kelembapan Kering : {_kering:.4f}")
print(f"Kelembapan Basah : {_basah:.4f}")

# Cell 4 - Aturan dan Nilai , z
z_min, z_max = 0, 30

# Rule 1: IF Cahaya Tinggi AND Tanah Kering THEN Penyiraman Panjang
1 = min(_cahaya_tinggi, _kering)
z1 = bertambah(1, *penyiraman_panjang())

# Rule 2: IF Cahaya Tinggi AND Tanah Basah THEN Penyiraman Pendek
2 = min(_cahaya_tinggi, _basah)
z2 = berkurang(2, *penyiraman_pendek())

# Rule 3: IF Cahaya Rendah AND Tanah Kering THEN Penyiraman Sedang
3 = min(_cahaya_rendah, _kering)
z3 = (berkurang(3, *penyiraman_sedang()) + bertambah(3, *penyiraman_sedang())) / 2  # Untuk fungsi segitiga

# Rule 4: IF Cahaya Rendah AND Tanah Basah THEN Penyiraman Pendek
4 = min(_cahaya_rendah, _basah)
z4 = berkurang(4, *penyiraman_pendek())

# Cetak aturan
print("\n=== ATURAN FUZZY ===")
for i, (, z) in enumerate([(1, z1), (2, z2), (3, z3), (4, z4)], 1):
    print(f"Rule {i} ->  = {:.4f}, z = {z:.2f}")

# Cell 5 - Defuzzifikasi
numerator = 1*z1 + 2*z2 + 3*z3 + 4*z4
denominator = 1 + 2 + 3 + 4
lama_penyiraman = numerator / denominator if denominator != 0 else 0

print(f"\n=== DEFUZZIFIKASI ===")
print(f"Lama Penyiraman (hasil defuzzifikasi): {lama_penyiraman:.2f} menit")

# Cell 6 - Visualisasi
x_cahaya = np.linspace(0, 10000, 200)
x_kelembapan = np.linspace(0, 100, 200)
x_output = np.linspace(0, 30, 200)

plt.figure(figsize=(15, 4))

# Intensitas Cahaya
plt.subplot(1, 3, 1)
plt.plot(x_cahaya, [cahaya_rendah(x) for x in x_cahaya], label='Rendah')
plt.plot(x_cahaya, [cahaya_tinggi(x) for x in x_cahaya], label='Tinggi')
plt.axvline(cahaya, color='red', linestyle='--', label=f'Input = {cahaya} lux')
plt.title('Fungsi Keanggotaan Intensitas Cahaya')
plt.xlabel('Intensitas Cahaya (lux)')
plt.ylabel('Derajat')
plt.legend()
plt.grid(True)

# Kelembapan Tanah
plt.subplot(1, 3, 2)
plt.plot(x_kelembapan, [tanah_kering(x) for x in x_kelembapan], label='Kering')
plt.plot(x_kelembapan, [tanah_basah(x) for x in x_kelembapan], label='Basah')
plt.axvline(kelembapan, color='red', linestyle='--', label=f'Input = {kelembapan}%')
plt.title('Fungsi Keanggotaan Kelembapan Tanah')
plt.xlabel('Kelembapan Tanah (%RH)')
plt.ylabel('Derajat')
plt.legend()
plt.grid(True)

# Output Lama Penyiraman
plt.subplot(1, 3, 3)
plt.plot(x_output, [turun(x, *penyiraman_pendek()) for x in x_output], label='Pendek')
plt.plot(x_output, [turun(x, *penyiraman_sedang()) + naik(x, *penyiraman_sedang()) for x in x_output], label='Sedang')
plt.plot(x_output, [naik(x, *penyiraman_panjang()) for x in x_output], label='Panjang')
plt.axvline(lama_penyiraman, color='red', linestyle='--', label=f'Output = {lama_penyiraman:.2f} menit')
plt.title('Lama Penyiraman')
plt.xlabel('Lama Penyiraman (menit)')
plt.ylabel('Derajat')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np

# Data
data = {
    'Nama': ['Ali', 'Budi', 'Citra', 'Dina', 'Eko', 'Fira', 'Gilang'],
    'Nilai Tugas': [80, 75, 90, 85, 70, 95, 88],
    'Nilai UTS': [78, 82, 85, 80, 76, 88, 84],
    'Nilai UAS': [90, 85, 88, 92, 80, 95, 89]
}

# Buat DataFrame
df = pd.DataFrame(data)

# Tambahkan kolom Nilai Akhir
df['Nilai Akhir'] = 0.3 * df['Nilai Tugas'] + 0.3 * df['Nilai UTS'] + 0.4 * df['Nilai UAS']

# Ambil kolom Nilai Akhir
nilai_akhir = df['Nilai Akhir']

# Hitung statistik deskriptif
mean = np.mean(nilai_akhir)
median = np.median(nilai_akhir)
std_dev = np.std(nilai_akhir)

# Tampilkan hasil
print("Rata-rata (Mean):", round(mean, 2))
print("Median:", round(median, 2))
print("Standar Deviasi (Std):", round(std_dev, 2))
```
